## 4. Filter the dataframe with only this columns¶

```python
covid_updated=covid.filter(items=['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
covid_updated.head()
```

## 5. Data Cleaning

### a. Remove all duplicates observations

```python
covid_duplicated=covid.copy()
```

```python
covid_duplicated=covid_duplicated.set_index("continent")
covid_duplicated.duplicated().sum()
```

```python
covid_duplicated.shape
```

```python
covid_duplicated=covid_duplicated.drop_duplicates()
```

```python
covid_duplicated.shape
```

### b. Find missing values in all columns

```python
covid.isnull()
```

### c. Remove all observations where continent column value is missing¶

```python
covid_missing_continent=covid.copy()
```

```python
covid_missing_continent["continent"].isnull().sum()
```

```python
covid_missing_continent["continent"].shape
```

```python
covid_missing_continent=covid_missing_continent.dropna(subset=["continent"])
```

```python
covid_missing_continent["continent"].shape
```

### d. Fill all missing values with 0

```python
covid=covid.fillna(0)
```

# Case Study 2
## 6. Date time format :
### a. Convert date column in datetime format using pandas.to_datetime

```python
covid.dtypes["date"]
```

```python
covid["date"]=pd.to_datetime(covid["date"])
```

```python
covid.dtypes["date"]
```

### b. Create new column month after extracting month data from date column.

```python
covid["month"]=covid["date"].dt.month
covid.dtypes
```

## 7. Data Aggregation:
### a. Find max value in all columns using groupby function on 'continent' column
### b. Store the result in a new dataframe named 'df_groupby'.


```python
df_groupby=covid.groupby("continent").max()
```

```python
df_groupby
```

```python
df_groupby=df_groupby.reset_index()
```

```python
df_groupby
```

## 8. Feature Engineering :
### a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby["total_deaths_to_total_cases"]=(df_groupby["total_deaths"]/df_groupby["total_cases"])*100
```

```python
df_groupby
```

## 9. Data Visualization :
### a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

```python
sns.displot(df_groupby["gdp_per_capita"])
```

### b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'¶

```python
sns.jointplot(data=df_groupby,x="total_cases",y="gdp_per_capita",kind="scatter")
```


### c. Plot Pairplot on df_groupby dataset.

```python
sns.pairplot(data=df_groupby)
```

### d. Plot a bar plot of 'continent' column with 'total_cases' .

```python
sns.catplot(data=df_groupby,x="continent",y="total_cases",kind="bar")
```
